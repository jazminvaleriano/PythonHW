<h3 style="text-align:left;">Exercise 1 - Audio signal<span style="float: right">5 points</span></h3>

In this exercise you'll work with continuous audio signals. First we load and plot the audio signals:

```python
# Import dependencies
import time
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy.io.wavfile import read

fs_bird, sound_bird = read('./data/bird.wav')
fs_leaves, sound_leaves = read('./data/leaves.wav')

# Plot sounds in different subplots
plt.figure(figsize=(16,6))
plt.subplot(1, 2, 1)
plt.plot(sound_bird)
plt.title("Bird sound")
plt.subplot(1, 2, 2)
plt.plot(sound_leaves)
plt.title("Leaves sound")
plt.show()

# Play sounds
time.sleep(1)
sd.play(sound_bird, fs_bird)
time.sleep(1)
sd.play(sound_leaves, fs_leaves)
time.sleep(1)
```

1. Double the amplitude of the second audio signal (leaves) using *numpy*. Then plot the amplified signal using *matplotlib*. (*1 point*)

```python
###
# Amplify the original leaves sound by 2
amplified_leaves = 2 * sound_leaves

# Plotting the original signal using subplot and plot functions of matplotlib.pyplot
plt.figure(figsize=(16,6))
plt.subplot(1, 2, 1)
plt.plot(sound_leaves)
plt.title("Original Leaves sound")

# Plotting amplified signal using subplot and plot
plt.subplot(1, 2, 2)
plt.plot(amplified_leaves)
plt.title("Amplified Leaves sound")

# Using plt.show for displaying the plots
plt.show()
###
```

2. The two sounds don't have the same length (number of samples). Print the length of the two sounds and create a new leaves sound as long as the birds sound. Print again the lenghts to make sure they match. (Use the amplified leaves sound previously created). (*1 point*)

```python
###
# Print the length of the original sounds
print("Length of bird sound:", len(sound_bird))
print("Length of leaves sound:", len(sound_leaves))

# Create a new sound from the original amplified leaves, but with the length of the bird sound: 
amplified_leaves_copy = amplified_leaves[0:len(sound_bird)].copy()
print("Length of new (amplified) leaves sound:", len(amplified_leaves_copy))

###
```

3. Create a silence period (0 amplitude) of the same duration of the two sounds, using *numpy*. (*1 point*)

```python
###
# Create a silence period the same length as birds sound
import numpy as np
silence= np.zeros(len(sound_bird))
len(silence)
###
```

4. Finally, concatenate the diffent sounds into one. The order should be: silence, birds, silence, leaves, silence (use the amplified and cut leaves sound). Use *numpy* and plot the results. (*2 points*)

```python
###
concatenated_sounds = np.concatenate((silence, sound_bird, silence, amplified_leaves_copy, silence))
plt.figure(figsize=(12, 6))
plt.plot(concatenated_sounds)
plt.title("Concatenated Sound")
plt.show()

###
```

In this exercise, you'll work with an image, i.e. an n-dimensional matrix data. First, we load and plot the image:

```python
import matplotlib.pyplot as plt
import numpy as np

im = plt.imread('./data/python.bmp')
im = np.array(im)

plt.figure(figsize=(4, 4))
plt.imshow(im)
plt.show()
```

The image is a 3-dimensional array, where the 1st and 2nd dimensions represent positions on the Y (rows) and X (columns) axes and the 3rd saturation values between 0 and 255 for that specific position, as [red, green, blue].

1. Set the pixels on the 50th to 60th row and and 200th to 210th column to green, and displays the resultant image. The data-type of the array should be an 8-bit unsigned integer. (*1 point*)

**NB!** 
- From the 50th row, up to and including the 60th; From the 200th column, up to and including the 210th.
- Keep in mind that the first dimension is related to the usual Y axis, and the second dimension to the usual X axis.

```python
###
import matplotlib.pyplot as plt
import numpy as np

# Reading the image using plt.imread
im = plt.imread('./data/python.bmp')

# Converting the image to a NumPy array
im = np.array(im)
print("Data type of the array: ", im.dtype)

# Setting pixels on the 50th to 60th row and 200th to 210th column to green
im[50:61, 200:211, :] = [0, 255, 0]  # RGB values for green, red and blue are set to 0

# Creating a figure with a size of 4x4 inches
plt.figure(figsize=(4, 4))

# Displaying the updated image using the imshow function
plt.imshow(im)

# Showing the plot
plt.show()###
```

2. The image created in point 1 will be the one you use in the rest of the exercise. Make a copy of the top half of the image, flip it along the 2nd axis using the appropriate numpy method, and plot the result. (*1 point*)

```python
###

# Make a copy of the top half of the image, im.shape[0] extracts the size of the array along the vertical axis, 
# //2 calculates its mid point
# The colons represent all values of x, and a step size 1. 

top_half = im[:im.shape[0] // 2, :, :].copy()

# Flip the copied top half along the 2nd axis (horizontally) using NumPy's flip function. 
flipped_top_half = np.flip(top_half, axis=1)

#As per np manual, "along the second axis" means to flip horizontally: https://numpy.org/doc/stable/reference/generated/numpy.flip.html

# Create a figure with a size of 4x4 inches
plt.figure(figsize=(4, 4))

# Display the original top half
plt.subplot(1, 2, 1)
plt.imshow(top_half)
plt.title("Original Top Half")

# Display the flipped top half
plt.subplot(1, 2, 2)
plt.imshow(flipped_top_half)
plt.title("Flipped Top Half")

# Show the plot
plt.show()###
```

3. Make a copy of the bottom half of your image, combine it together with the first half you flipped in point 2, and display the result. Also plot your original image and check whether it has changed. (*1 point*)

```python
###

# Make a copy of the bottom half of the image
# Extracts rows starting from the midpoint to the end of array along the first axis
bottom_half = im[im.shape[0] // 2:, :, :].copy()

# Concatenate is used for combining the flipped top half with the bottom half
combined_image = np.concatenate([flipped_top_half, bottom_half], axis=0)

# Create a figure with a size of 12x4 inches, allows for decent display of the three plots
plt.figure(figsize=(12, 4))

# Display the original image
plt.subplot(1, 3, 1)
plt.imshow(im)
plt.title("Original Image")

# Display the flipped top half
plt.subplot(1, 3, 2)
plt.imshow(flipped_top_half)
plt.title("Flipped Top Half")

# Display the bottom half
plt.subplot(1, 3, 3)
plt.imshow(bottom_half)
plt.title("Bottom Half")

# Show the plot
plt.show()

# Create a figure with a size of 4x4 inches for the combined image
plt.figure(figsize=(4, 4))
plt.imshow(combined_image)
plt.title("Combined Image")
plt.show()###
```

```python
###
# Make a shallow copy (view) of the top half of the image
# Making changes to this will reflect on im as well. 
shallow_copy_top_half = im[:im.shape[0] // 2, :, :]

# Change every black pixel to a green pixel in the shallow copy
# axis=-1 refers to the last axis (where RGB colors are stored)
# black_pixels is a boolean in which TRUE represents positions of black
black_pixels = (shallow_copy_top_half == [0, 0, 0]).all(axis=-1)
# Green is set to 255, maximum intensity
shallow_copy_top_half[black_pixels] = [0, 255, 0]

# Create a figure with a size of 12x4 inches
plt.figure(figsize=(12, 4))

# Display the original image
plt.subplot(1, 3, 1)
plt.imshow(im)
plt.title("Original Image")

# Display the shallow copy (colored top part)
plt.subplot(1, 3, 2)
plt.imshow(shallow_copy_top_half)
plt.title("Shallow Copy (Colored Top Part)")

# Display the base of the shallow copy 
plt.subplot(1, 3, 3)
plt.imshow(shallow_copy_top_half.base)
plt.title("Base of Shallow Copy")

# Show the plot
plt.show()###
```

<h3 style="text-align:left;">Exercise 3 - Tabular data<span style="float: right">3 points</span></h3>

In the following you'll work with tabular data, i.e. data related to multiple observations. The dataset consists of synthetic data on monthly average precipitations in 4 different countries: Switzerland, Italy, France and Germany. Let's first load the dataset and assign the variables:

```python
import numpy as np

precip = np.load("./data/precip.npy", allow_pickle=True)

precip[:, 0]  # 12 data points for Switzerland (one per month)
precip[:, 1]  # 12 data points for Italy (one per month)
precip[:, 2]  # 12 data points for France (one per month)
precip[:, 3]  # 12 data points for Germany (one per month)
```

1. Use built-in numpy methods to obtain an array with the maximum precipitation for each country, and the average precipitation *across* the five countries for each month. (*1 point*)

```python
###
precipitation = np.load("./data/precip.npy")

# Get the maximum precipitation for each country using the max function
max_precipitation_per_country = np.max(precip, axis=0)

# Get the average precipitation across the five countries for each month using the mean function
avg_precipitation_per_month = np.mean(precipitation, axis=1)

print("Maximum precipitation per country:", max_precipitation_per_country)
print("Average precipitation per month across countries:", avg_precipitation_per_month)###
```

2. Obtain the maximum precipitation of the entire dataset. Next, use a numpy method to find out at which row and column in the dataset you would find this maximum precipitation at. (*2 points*)

```python
###
#Find the max precipitation value of the entire dataset
max_value=np.max(precip)

# Find the index of the max precipitation
max_index=np.argmax(precip)

# Convert the flat index to row and column indices
max_row, max_col = np.unravel_index(max_index, precip.shape)

print("Maximum precipitation of entire dataset: ",max_value)
print("Located at row: ",max_row,", column: ",max_col)

###
```
